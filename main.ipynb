```python
# Simple Recursion
def show(n):
    if n==0:
        return
    print(n)
    show(n-1)

show(5)
```

```python
# Inheritance 
# One class acquires properties of another class

class parent:
    def house(self):
        print("Parent House")

class child(parent):
    def bike(self):
        print("child bike")

c = child()
c.house()
c.bike()
```

```python
# Polymorphism
# One function - many forms
class A:
    def show(self):
        print("class A")
class B(A):
    def show(self):
        print("Class B")
    
obj = B()
obj.show()
```

```python
# Encapsulation
# Binding data and methods together

class Bank:
    def __init__(self):
        self.__balance = 1000
    def show_balance(self):
        print(self.__balance)

b = Bank()
b.show_balance()
```

```python
#Class: Blueprint of object
#Object: Instance of class
#Constructor: Initializes data
#self: Current object
#Inheritance: Parentâ€“child relation
#Polymorphism: One name many forms
#Encapsulation: Data hiding
```
